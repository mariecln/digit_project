# Lecture des données

```python
f = open("digits.svm", "r")
data_as_string = f.read().splitlines()
```

```python
n = int(data_as_string.pop(0)) #Lecture de la première ligne, et on la vire
print(n)
nb_car = int(data_as_string.pop(0)) # Lecture de la nouvelle première ligne, et on la vire
print(nb_car)
#data as string c'est un tableau avec chaque élément qui correspond à une ligne, faites des prints si c'est pas clair
```

```python
X = []
Y = [] 
for i in range(n):
    Y.append(int(data_as_string[i][0])) 
    X.append(data_as_string[i].split(" ")[1:])
    X[i] = [float(val) for val in X[i]] #On convertit les chaînes de caracrètres en float

print("X[1] : ", X[1])
print("Longueur X[1] : ", len(X[1]))
print("Y[1] :", Y[1])
print("Taille X : ", len(X))
print("Taille Y : ", len(Y))
```

1000 lignes de données, séparation en 80-20% avec tirage aléatoire

```python
training_data = X[0:800]
training_labels = Y[0:800]

test_data = X[800:]
test_labels = Y[800:]

print(len(training_data), len(test_data))
```

```python
def normeEuclidienne(x):
    res=0
    for i in range(len(x)):
        res+=(x[i])*(x[i])
    return math.sqrt(x)

def distanceEuclidienne(x,y):
    res=0
    for i in range(len(x)):
        res+=(x[i]-y[i])*(x[i]-y[i])
    return math.sqrt(res)
```

```python
distanceEuclidienne(test_data[1], training_data[1])
```

```python
def predictData(training_data,training_labels , test_sample, K):
    similarity = [] #Un vecteur qui sera de taille 800 pour calculer la similarité avec tous les échantillons
    
    for i in range(len(training_data)):
        similarity.append(distanceEuclidienne(training_data[i], test_sample))
    
    sorted_indexes = sorted((e,i) for i,e in enumerate(similarity)) #On trie le tableau et on récupère les indices
    print("sorted indexes : ", sorted_indexes[:3],"...")
    
    nearest_neighbors = {}
    nearest_neighbors_count = {}
    
    for i in range(K):
        
        neighbor = sorted_indexes[i] 
        print(neighbor, training_labels[neighbor[1]])
        
        index = neighbor[1]
        distance = neighbor[0]
        
        if training_labels[index] in nearest_neighbors.keys():
            nearest_neighbors[training_labels[index]] += distance
            nearest_neighbors_count[training_labels[index]] += 1
        else:
            nearest_neighbors[training_labels[index]] = distance
            nearest_neighbors_count[training_labels[index]] = 1
     
    print("Nearest Neighbors : ", nearest_neighbors)
    for key in nearest_neighbors.keys():
        nearest_neighbors[key] = nearest_neighbors[key]/nearest_neighbors_count[key] 
    
    print("Nearest Neighbors : ", nearest_neighbors)
    
    best_prediction = None
    best_score = 9999999
    for key,val in nearest_neighbors.items():
        if val < best_score:
            best_score = val
            best_prediction = key
    return best_prediction
```

```python
print("Bonne réponse : ", test_labels[7])
predictData(training_data, training_labels, test_data[7], 5)
```

```python
predictions = []
for sample in test_data:
    predictions.append(predictData(training_data, training_labels, sample, 5))
```
